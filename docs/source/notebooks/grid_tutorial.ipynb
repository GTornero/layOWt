# Tutorial 1: Generating a `Grid`

LayOWt is a very simple package to use for layout design, taking care of the more mundane and manual tasks for you.

The most common way to define a wind turbine layout is with a uniformly spaced grid of points. The definition and creation of such points is taken care of by the aptly named `Grid` object.


## 1. Import the `grid` object from `layowt.grids`

```python
from layowt.grids import Grid
```

## 2. Defining a `Grid`
`Grid` objects in LayOWt can be defined by many input arguments when initialised. You can read the full API documentation for this object under layowt.grids.grid.
The following list describes the arguments and their meanings:

1. n_rows: Number of rows of points in the `Grid`
2. n_cols: Number of columns of points in the `Grid`
3. row_step: Distance between rows in the `Grid`. When used alongside the `scale` argument the distance between rows will be row_step*scale.
4. col_step: Distance between columns in the `Grid`. When used alongside the `scale` argument the distance between columns will be col_step*scale.
5. row_offset: Offset every other row horizontally by half row_step*scale to create diamond shaped grids.
6. col_offset: Offset every other column vertically by half col_step*scale to create diamond shaped grids.
7. angle: Angle of orientation of the `Grid` columns in degrees. Clockwise convention with zero at North to match the Oceanographic convention.
8. x_shear: Horizontal shear angle of the `Grid` in degrees. Defines the clockwise angle of the columns from the vertical axis.
9. y_shear: Vertical shear angle of the `Grid` in degrees. Defines the anticlockwise angle of the rows from the horizontal axis.
10. origin: (x,y) coordinates of the centroid of the `Grid`.
11. scale: Factor by which to convert the non-dimensional row_step and col_step spacings into real space. Rotor diameter will be used in these tutorials as the `Grid` scale.

### 2.1 Default `Grid`
A default `Grid` can be initialised without passing any arguments. It can also be plotted as a matplotlib Figure with its `plot` method.

```python
grid = Grid()
grid.plot()
```

### 2.2 Changing the spacing of the `Grid`
The `Grid` spacing can be defined as a number of rotor diameters (in this case 250m) as per the following example:

```python
grid = Grid(n_rows=20, n_cols=20, row_step=10, col_step=6, scale=250)
grid.plot()
```

### 2.3 Setting the `Grid` orientation
The orientation of the `Grid` can be set with the `angle` argument.
Consider the resulting grid using the same input arguments as above but with a value passed for `angle`:

```python
grid = Grid(n_rows=20, n_cols=20, row_step=10, col_step=6, angle=45, scale=250)
grid.plot()
```

### 2.4 Visualizing shears and offsets
Offsets and shears are not as simple to visualize as spacings and rotations.

The following examples show the effect of the offset and shear arguments:

.. note::
    row_offset and col_offset cannot be enabled at the same time.

- Offset:

```python
grid = Grid(n_rows=10, n_cols=10, row_step=10, col_step=6, scale=250, row_offset=True)
grid.plot()
```

- Shear:

```python
grid = Grid(n_rows=10, n_cols=10, row_step=10, col_step=6, scale=250, x_shear=30)
grid.plot()
```

## 3. Extracting `Grid` Geometries & Coordinates
If we wish to extract geometric or coordinate information of the `Grid` we just created, we can do so with the following methods and properties:

- to_points: returns a list of shapely Point geoemtries representing each point within the `Grid`
- to_multipoint: returns a shapely MultiPoint object representing the full `Grid` of points
- coords: property of the `Grid` as an array of (x, y) coordinate pairs for all points in the `Grid`
- x: property of the `Grid` as a 1D array of the x-coordinates for all points in the `Grid`
- y: property of the `Grid` as a 1D array of the y-coordinates for all points in the `Grid`

```python
grid = Grid(n_rows=3, n_cols=3)
grid.to_multipoint()
```

```python
print(grid.coords)
```

```python
print(grid.x)
```
